```python
import requests
from bs4 import BeautifulSoup
import requests

import re
import string 

def clean_sentence(text):
    text = re.sub(r'\[\d+\]', " ", text)
    text = re.sub(r"[\([{})\]]", " ", text)
    text = re.sub("[\(\[].*?[\)\]]", "", text)
    output= "".join([i for i in text])
    return output


corpus = ""
titles = list()
page_corpus_list = list()
for x in range(10000):
    URL = "https://en.wikipedia.org/wiki/Special:Random"
    page = requests.get(URL)
    
    soup = BeautifulSoup(page.content, 'html.parser')
    page_text = soup.find_all('p')
    title = soup.title.text
    titles.append(title)

    page_corpus = " "
    for i in range(len(page_text)):
        try:
            text = page_text[i].get_text()
            page_corpus += clean_sentence(text)
            page_corpus = page_corpus.strip()
        except:
            page_corpus = " "
    page_corpus_list.append(page_corpus)
    corpus += page_corpus
    if x%25 == 0:
        print(x)

with open('Wikipedia Corpus Cleaned.txt', 'w') as f:
    f.write(corpus)
```
